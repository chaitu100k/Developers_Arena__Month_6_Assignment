# Cleaning & EDA

Basic cleaning and exploratory data analysis.

```python
DATA_PATH = "C:/Users/Asus/Documents/Developers_Arena_Internship_Files/Month_6_Internship_files/traffic_flow_project/traffic_flow_project/data/synthetic_traffic_dataset.csv"
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(DATA_PATH, parse_dates=['timestamp'])
print(df.info())

# Basic cleaning
print('rows before:', len(df))
df = df.dropna()
print('rows after dropna:', len(df))

# Simple plot: vehicle_count over time (first sensor)
try:
    s1 = df[df['sensor_id']==1]
    plt.figure(figsize=(8,3))
    plt.plot(s1['timestamp'], s1['vehicle_count'])
    plt.title('Vehicle count over time (sensor 1)')
    plt.xlabel('timestamp')
    plt.ylabel('vehicle_count')
    plt.show()
except Exception as e:
    print('Plot failed:', e)
```
